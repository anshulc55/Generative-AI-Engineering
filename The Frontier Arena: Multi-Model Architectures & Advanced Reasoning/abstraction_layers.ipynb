# LangChain vs. LiteLLM Abstractions

## Architect's Focus: Minimizing Technical Debt

In this lab, we stop talking and start coding. We will implement a unified query system using both **LiteLLM** and **LangChain** to compare their syntax, weight, and ease of use.

```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display

# Framework 1: LiteLLM (The Lightweight Proxy)
from litellm import completion

# Framework 2: LangChain (The Heavyweight Ecosystem)
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

load_dotenv(override=True)
print("✅ Frameworks Loaded.")
```

## 1. LiteLLM Implementation (The OpenAI Standard)
LiteLLM uses the exact same syntax as the OpenAI library. The only difference is you prefix the model name with the provider.

```python
def litellm_query(model, prompt):
    """Zero-learning-curve abstraction."""
    print(f"🚀 LiteLLM -> {model}")

    response = completion(
        model=model, 
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content

prompt = "Why is 'Standardization' important in AI APIs?"

# Testing across different providers with ZERO code change
models = ["openai/gpt-4o-mini", "deepseek/deepseek-chat"]

for m in models:
    result = litellm_query(m, prompt)
    display(Markdown(f"**{m}:** {result}"))
```

## 2. LangChain Implementation (The Object-Oriented Way)
LangChain treats models as 'Objects' that you instantiate and then 'Invoke'.

```python
def langchain_query(model, prompt):
    """Object-oriented abstraction."""
    print(f"📦 LangChain -> {model}")
    llm = ChatOpenAI(model=model)
    
    # Note: Requires wrapping prompt in a Message object for best practice
    message = [HumanMessage(content=prompt)]
    response = llm.invoke(message)
    return response.content

result = langchain_query("gpt-4o-mini", prompt)
display(Markdown(f"**LangChain Result:** {result}..."))
```

## 3. The Comparison: Error Handling
Let's see which library provides better debug info when we pass a garbage model name.

```python
def test_error(framework_func):
    try:
        framework_func("invalid-model-name-123", "Hello?")
    except Exception as e:
        print(f"❌ ERROR CAUGHT: {str(e)[:150]}...")

print("Testing LiteLLM Error:")
test_error(litellm_query)

print("\nTesting LangChain Error:")
test_error(langchain_query)
```
