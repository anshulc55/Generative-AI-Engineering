# Decentralizing AI: Local Inference with Ollama and OpenRouter

As an Architect, you must always plan for the **"Privacy-Cost-Latency"** triangle. 
1. **Cloud AI:** High power, but high cost and zero privacy.
2. **Local AI:** Zero cost, 100% privacy, but limited by your laptop's RAM.

Today, we break free from the Big Tech "gatekeepers." We will teach you how to run **Llama 3** and **Mistral** locally on your machine using **Ollama**, and how to use **OpenRouter** as a failover bridge to hundreds of other open-weights models.

```python
import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display

load_dotenv(override=True)

# 1. Initialize our Decentralized Clients
# Ollama runs on port 11434 by default
client_local = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama" # Ollama doesn't need a real key, but the library requires one
)

# OpenRouter provides one API for all open-source models
client_openrouter = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.getenv("OPENROUTER_API_KEY")
)

print("🔓 Decentralized Clients Initialized!")
```

## 1. Running Local: The Ollama Fortress

**Pre-requisite:** You must have the Ollama app running on your machine. 

```python
def query_local(prompt, model="smollm:1.7b"):
    try:
        print(f"Querying LOCAL model: {model}...")
        response = client_local.chat.completions.create(
            model=model,
            messages=[{"role": "user", "content": prompt}]
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"❌ Local Inference Error: {e}. (Is Ollama running?)"

prompt = "Write a 2-line poem about data privacy."
answer = query_local(prompt)
display(Markdown(f"### Local Output:\n{answer}"))
```

## 2. The Bridge: OpenRouter

Sometimes your local machine isn't powerful enough for a "Reasoning" task. Instead of going back to OpenAI, we can use **OpenRouter** to access massive open-source models like **Llama-3-70B** or **Qwen-2.5**.

```python
def query_openrouter(prompt, model="openai/gpt-oss-120b:free"):
    print(f"🌐 Querying OPENROUTER model: {model}...")
    response = client_openrouter.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content

answer = query_openrouter("Explain the concept of 'Stock P/E Ratio' in 2 sentence.")
display(Markdown(f"### OpenRouter Output:\n{answer}"))
```

## 3. The "Local-First" Hybrid Pattern

An expert Architect builds systems that are resilient. We try to solve the problem **locally** first (free and private). If the local server is down or the task is too hard, we failover to the **Cloud**.

```python
def smart_query(prompt):
    # 1. Try Local First
    local_ans = query_local(prompt)
    
    if "❌" not in local_ans:
        print("✅ Resolved Locally (Free & Secure)")
        return local_ans
    
    # 2. Fallback to OpenRouter
    print("⚠️ Local failed. Falling back to OpenRouter cloud...")
    return query_openrouter(prompt)

result = smart_query("What is the capital of Mars?")
display(Markdown(result))
```
