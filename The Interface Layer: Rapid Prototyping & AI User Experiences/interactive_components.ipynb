# Interactive Components: State, Callbacks, and Public Deployment

A professional AI application is more than just a Chatbox. It is a system of **Events**. 
1. **Callbacks:** Functions that trigger automatically when a user interacts (clicks, types, or slides).
2. **State:** The "Short-Term Memory" of your UI. It allows your app to remember data across multiple clicks without a database.
3. **Deployment:** The ability to instantly share your creation with stakeholders for feedback.

```python
import gradio as gr
from litellm import completion
from dotenv import load_dotenv
import os

load_dotenv(override=True)

print("⚡ Interactive Environment Ready!")
```

## 1. Managing State: The Session Memory

In a standard function, variables are lost once the function finishes. In Gradio, we use `gr.State()` to keep data alive for the duration of a user's session. 

**Example:** An AI Note-Taker that accumulates ideas before generating a final summary.

```python
def add_note(current_note, session_notes):
    """Adds a note to the state and returns the updated list."""
    session_notes.append(current_note)
    return "", f"Total Notes: {len(session_notes)}", session_notes

def finalize_research(session_notes):
    """Uses the state to generate a report."""
    if not session_notes:
        return "No notes found!"
    
    all_notes = "\n- ".join(session_notes)
    prompt = f"Based on these research notes, write a professional summary:\n{all_notes}"
    
    response = completion(model="openai/gpt-4o-mini", messages=[{"role": "user", "content": prompt}])
    return response.choices[0].message.content

with gr.Blocks() as state_demo:
    # Initialize our Session State
    notes_state = gr.State([]) 
    
    gr.Markdown("# 📝 AI Research Architect")
    
    with gr.Row():
        note_input = gr.Textbox(label="Enter a Research Finding", placeholder="E.g., Competitor X uses RAG...")
        add_btn = gr.Button("Save Note", variant="secondary")
    
    status_text = gr.Label("Total Notes: 0")
    
    with gr.Accordion("Final Report", open=False):
        report_btn = gr.Button("Generate Summary", variant="primary")
        report_output = gr.Markdown()

    # CALLBACK LOGIC: Triggering events
    add_btn.click(
        fn=add_note, 
        inputs=[note_input, notes_state], 
        outputs=[note_input, status_text, notes_state]
    )
    
    report_btn.click(
        fn=finalize_research, 
        inputs=notes_state, 
        outputs=report_output
    )

print("✅ Stateful App Logic Defined.")
# state_demo.launch()
```

## 2. Public Deployment: The 'Share' Flag

As an Architect, you need to show your work to your boss or a client *now*. Gradio includes a built-in proxy that exposes your local laptop to a public `gradio.live` URL.

**⚠️ Security Warning:** Only use `share=True` for temporary demos. Anyone with the link can access your app and potentially consume your API credits.

```python
# To launch this and get a public URL for stakeholders:
state_demo.launch(share=True)

print("💡 Pro Tip: When you launch with share=True, Gradio gives you a 72-hour public link.")
```
