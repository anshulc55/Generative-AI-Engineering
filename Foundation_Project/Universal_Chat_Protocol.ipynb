<h2 style="color:#e67e22;">The Anatomy of an API Endpoint</h2>

Before we rely on automated libraries, we must understand the raw communication happening under the hood. An **Endpoint** is essentially a digital address where our request is processed.

When we talk to an LLM, we are sending a **JSON Payload** via an **HTTP POST** request. Let's construct a raw payload to see exactly what the model sees.

```python
from typing import Dict, Any, List

def construct_chat_payload(model: str, user_message: str) -> Dict[str, Any]:
    """
    Constructs a standard Chat Completions JSON payload.
    """
    return {
        "model": model,
        "messages": [
            {"role": "user", "content": user_message}
        ]
    }

test_payload = construct_chat_payload("gpt-4o-mini", "Explain in how many days Earth will be destroyed by the sun.")
print(f"Request Payload:\n{test_payload}")
```

```python
import os
import requests
from dotenv import load_dotenv

env_loaded = load_dotenv(override=True)

def execute_raw_inference(payload: Dict[str, Any]) -> Dict[str, Any]:
    """
    Executes an inference request using the requests library (Raw HTTP).
    """
    api_url = "https://api.openai.com/v1/chat/completions"
    headers = {
        "Authorization": f"Bearer {os.getenv('OPENAI_API_KEY')}",
        "Content-Type": "application/json"
    }
    
    response = requests.post(api_url, headers=headers, json=payload)
    return response.json()

raw_response = execute_raw_inference(test_payload)
#print(raw_response)
content = raw_response.get('choices', [{}])[0].get('message', {}).get('content')
print(f"Model Output:\n{content}")
```
