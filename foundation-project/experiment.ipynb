```python
# Check if we are in a Virtual Environment
in_venv = (sys.prefix != sys.base_prefix)

if in_venv:
    print("✅ SUCCESS: Running inside a UV Virtual Environment.")
else:
    print("WARNING: You are running in the Global System Python.")
    print("Please switch the Kernel to the project's '.venv'.")
```
