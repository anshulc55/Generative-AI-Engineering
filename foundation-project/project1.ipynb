```python
import os
from dotenv import load_dotenv
def validate_environment_setup() -> None:
    """
    Verifies that all required API keys are properly configured in the environment.
    """

    # Explicitly load the .env file and check if it was successful
    env_loaded = load_dotenv(override=True)

    key_configs = {
        "OPENAI_API_KEY": "sk-",
        "ANTHROPIC_API_KEY": "sk-ant-",
        "GEMINI_API_KEY": "AIza",
        "DEEPSEEK_API_KEY": "sk-"
    }

    print("--- Environment Configuration Check ---")
    for key, expected_prefix in key_configs.items():
        value = os.getenv(key)

        if not value:
            print(f"❌ {key:<20}: Not found in .env")
            continue
        
        if not value.startswith(expected_prefix):
            print(f"{key:<20}: Found, but expected prefix '{expected_prefix}'")
        elif value.strip() != value:
            print(f"{key:<20}: Found, but contains hidden whitespace")
        else:
            print(f"{key:<20}: Validated (ends with ...{value[-4:]})")

validate_environment_setup()
```

## Quick Preview: Your First API Call

```python
from openai import OpenAI

def quick_test_frontier_models():
    # 1. Test OpenAI
    print("--- Testing OpenAI ---")
    openai_client = OpenAI()
    response = openai_client.chat.completions.create(
        model="gpt-5-mini",
        messages=[
            {"role": "system", "content": "You are a grumpy 5-year-old."},
            {"role": "user", "content": "What is the capital of the moon?"}
        ]
    )
    print(response.choices[0].message.content)
    

    # 2. Test Gemini
    print("--- Testing Deepseek ---")
    gemini_client = OpenAI(
        base_url="https://api.deepseek.com/v1",
        api_key=os.getenv("DEEPSEEK_API_KEY")
    )
    response = gemini_client.chat.completions.create(
        model="deepseek-chat",
        messages=[
            {"role": "system", "content": "You are a grumpy 5-year-old."},
            {"role": "user", "content": "What is the capital of the moon?"}
        ]
    )
    print(response.choices[0].message.content)
quick_test_frontier_models()
```

### The Art of Prompt Engineering: System vs. User
To build effective AI applications, you must understand the two primary levers of control in a chat-based LLM API:

### 1. The System Prompt (The "Who")
Think of the System Prompt as the **personality, expertise, and boundary-setter**. It defines the model's persona and rules of engagement before the conversation even begins. 
- **Purpose:** Sets tone, formatting rules, language constraints, and "guardrails."
- **Impact:** It stays "behind the scenes" but influences every subsequent response.

### 2. The User Prompt (The "What")
This is the **specific task or question** provided by the user.
- **Purpose:** Provides the data to be processed or the question to be answered.
- **Impact:** It triggers the model to act based on the persona defined in the system prompt.

### One Query, Many Personalities
See how the **same User Prompt** (*"Explain why the sky is blue"*) changes based on different **System Prompts**:

| System Prompt Persona | LLM Response Style |
| :--- | :--- |
| **"You are a Nobel Prize winning Physicist."** | "The blue color of the sky is a result of Rayleigh scattering, where shorter wavelengths of sunlight are scattered more by atmospheric molecules..." |
| **"You are a pirate on the high seas."** | "Arrr, matey! The Great Blue is blue 'cause the tiny air bits be tossin' the sun's blue light all 'round the deck!" |
| **"You are a grumpy 5-year-old."** | "Because it is! Stop asking me questions! I want candy!" |
| **"You are a minimalist poet."** | "Sun hits air / Blue scatters / Heaven glows." |

**Key Takeaway:** By mastering the System Prompt, you can transform a general-purpose model into a specialized business tool.
