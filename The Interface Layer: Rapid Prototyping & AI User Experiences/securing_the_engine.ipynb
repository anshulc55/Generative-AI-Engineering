# Securing the Engine: Authentication & Direct LLM Integration

In a production environment, your AI application is a **Financial Asset**. Every request costs money. Today, we secure the "Front Gate" of our application using **Basic Authentication** and wire the UI directly to our **OpenAI / LiteLLM Backend**.

We move from an "Open Lab" to a "Secure Enterprise Portal."

```python
import gradio as gr
from litellm import completion
from dotenv import load_dotenv
import os

load_dotenv(override=True)

print("🔐 Security Framework Loaded.")
```

## 1. The "Lock on the Door": Gradio Authentication

Gradio provides a built-in `auth` parameter that intercepts all requests before they reach your logic. As an Architect, this is your first line of defense.

```python
def secure_ai_logic(prompt):
    """Our Production-Ready Backend"""
    response = completion(
        model="openai/gpt-4o-mini", 
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content

# WIRING THE SECURE PORTAL
with gr.Blocks(theme=gr.themes.Monochrome()) as secure_portal:
    gr.Markdown("# 🏦 Enterprise AI Secure Portal")
    gr.Markdown("This interface is locked behind the Corporate Identity Provider (Simulated).")
    
    user_input = gr.Textbox(label="Secure Command", placeholder="Enter your internal query...")
    output_area = gr.Markdown()
    
    user_input.submit(fn=secure_ai_logic, inputs=user_input, outputs=output_area)

print("✅ Secure Portal Defined.")

# THE ARCHITECT'S LOCK:
# To launch with security, we pass a list of (username, password) tuples.
authorized_users = [
    ("architect", "gemini2026"),
    ("admin", "prod-safe-99")
]

secure_portal.launch(auth=authorized_users)
```

## 2. Direct Backend Wiring

Now, we wire our secure UI to a more complex **Multi-Model Backend**. This pattern allows the user to select their "Reasoning Level" while staying within the secure perimeter.

```python
def multi_model_backend(prompt, model_choice):
    """
    Directly wiring the UI selection to the LiteLLM engine.
    """
    model_map = {
        "Fast (GPT-4o-mini)": "openai/gpt-4o-mini",
        "Deep (DeepSeek-R1)": "deepseek/deepseek-reasoner"
    }
    
    selected_model = model_map.get(model_choice, "openai/gpt-4o-mini")
    
    response = completion(model=selected_model, messages=[{"role": "user", "content": prompt}])
    return response.choices[0].message.content

with gr.Blocks() as pro_portal:
    gr.Markdown("# 🛡️ Pro-Architect: Authenticated Hub")
    
    with gr.Row():
        model_selector = gr.Dropdown(
            choices=["Fast (GPT-4o-mini)", "Deep (DeepSeek-R1)"], 
            value="Fast (GPT-4o-mini)", 
            label="Select Reasoning Engine"
        )
        query_input = gr.Textbox(label="System Prompt")
    
    submit_btn = gr.Button("Execute Secure Query", variant="primary")
    result_view = gr.Markdown()

    submit_btn.click(
        fn=multi_model_backend, 
        inputs=[query_input, model_selector], 
        outputs=result_view
    )

print("🚀 Professional Authenticated Hub Defined.")
pro_portal.launch(auth=authorized_users)
```
