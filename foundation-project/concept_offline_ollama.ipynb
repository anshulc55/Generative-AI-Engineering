# Concept: The Offline Brain – Connecting the OpenAI SDK to Ollama

--- 
In a production environment, you don't always need a massive, cloud-based model. Sometimes, the best architectural choice is a **Small Language Model (SLM)** running locally. This approach provides **zero-cost inference**, **total privacy**, and **low-latency responses**.

In this lesson, we will explore how to use specialized SLMs for three distinct tasks: General Chat, Code Writing, and Data Analysis. We'll use the OpenAI SDK as our universal interface to keep our code clean and portable.

## 2. The Multi-Model Switchboard

**Architect's Note:** Notice how the `base_url` remains the same for every model. Ollama acts as a 'Local Model Gateway'.

````python
from openai import OpenAI

client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)

def query_local_llm(model, prompt, system_prompt="You are a helpful assistant."):
    try:
        response = client.chat.completions.create(
            model=model,
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": prompt}
            ],
            temperature=0.3 # Lower temperature for more factual analysis
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"Error with {model}: {e}"
````

## 3. Example 1: General Chat (SmolLM)

**Model:** `smollm:1.7b` (1.7B)
**Use Case:** Quick Q&A, brainstorming, and mobile-ready assistant tasks.

SmolLM is a family of models from Hugging Face designed to be tiny enough to run on your phone while being trained on high-quality synthetic data to maintain coherence.

````python
prompt = "Give me a simple 3-step guide to starting a fitness routine for a busy professional."
print(f"SmolLM (General Chat):\n{query_local_llm('smollm:1.7b', prompt)}")
````

## 4. Example 2: Code Writing (DeepSeek Coder)

**Model:** `deepseek-coder:1.3b` (1.3B)
**Use Case:** Writing Python scripts, debugging, and explaining algorithms.

DeepSeek Coder is optimized specifically for programming. Despite its tiny size, it understands syntax and logic remarkably well.

````python
prompt = "Write a Python function to print 1 to 20 Table."
system = "You are an expert Python developer. Provide clean code with comments."
print(f"DeepSeek Coder (Programming):\n{query_local_llm('deepseek-coder:1.3b', prompt, system)}")
````

## 5. Example 3: Data Analysis & Summarization (Qwen2)

**Model:** `qwen2:1.5b` (1.5B)
**Use Case:** Extracting key facts from long text and analyzing sentiment.

Qwen2 is an extremely fast and precise model, perfect for 'skimming' through data to extract insights.

````python
news_article = """
Tech Giant 'CyberCore' announced today a major breakthrough in quantum networking, 
claiming to have achieved stable entanglement over 100 kilometers of existing fiber optic cable. 
Industry experts suggest this could pave the way for a secure 'Quantum Internet' by 2030, 
though critics point to the massive cooling infrastructure required as a significant hurdle.
"""
prompt = f"Analyze this news snippet. Extract: 1. The main event, 2. The key timeline, and 3. The main challenge.\n\nArticle: {news_article}"
system = "You are a news analyst. Be concise and use bullet points."

print(f"Qwen2 (Analysis):\n{query_local_llm('qwen2:1.5b', prompt, system)}")
````
